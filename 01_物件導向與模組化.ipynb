## 類似表格觀念

## 類別與物件
定義類別 (class)
* 定義建構式 (官方稱為 constructor，也可稱為 initializer)：
方便建立並初始化物件
* 定義屬性 (官方稱為 attribute，也可稱為 instance variable)

建立物件 (object)
* 呼叫建構式會建立實例 (instance)
* book = Book("111", "Python", 500, "Paul")
```
     ┌────┐   ┌───────────────────────────────────────┐
book │0x86├─> │ "111"  |  "Python"  |  500  |  "Paul" │
     └────┘   └───────────────────────────────────────┘
    (物件變數)             (實例, instance)

```

````python
# 定義類別
class Book:
    # 定義建構式。self 代表現行物件，也可更改為其他名稱
    def __init__(self, isbn, name, price, author):
        # Python 屬性定義在建構式內：
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author


def main():
    # 呼叫建構式以建立並初始化物件
    book = Book("111", "Python", 500, "Paul")
    # 修改屬性值
    book.author = "Peter"
    print("book.isbn:", book.isbn)
    print("book.name:", book.name)
    print("book.price:", book.price)
    print("book.author:", book.author)


main()
````

## 方法
類別內定義的函式稱作方法 (method)，主要提供給物件呼叫

````python
class Book:
    def __init__(self, isbn, name, price, author):
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    # self 代表現行物件。類別內不能重複定義方法 (名稱不可相同)
    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")

    def sale_price(self, discount=0.0):
        return self.price * (1 - discount)


def main():
    book1 = Book("111", "Python", 500, "Paul")
    book1.show()
    print("售價 (打八折):", book1.sale_price(0.2))


main()
````

## 類別變數
* 類別變數 (class variable) 是直接定義在類別內、但在方法之外的變數
* 類別變數由所有物件共同分享，而非獨屬於某個物件
* 一般會搭配類別方法 (class method) 來存取類別變數

````python
class Book:
    # 類別變數
    count = 0

    # 定義類別方法，需要加上 @classmethod。cls 代表現行類別，也可更改為其他名稱
    @classmethod
    def show_count(cls):
        print("cls.count:", cls.count)

    def __init__(self, isbn, name, price, author):
        Book.count += 1
        # self.xxx 會宣告實例變數
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")


def main():
    # 使用類別名稱存取類別變數
    print(Book.count)

    print("-" * 30)
    book1 = Book("111", "Python", 500, "Paul")
    book1.show()
    book2 = Book("222", "C++", 400, "Cindy")
    book2.show()
    print("-" * 30)

    # 使用類別名稱呼叫類別方法
    Book.show_count()


main()
````

## 靜態方法 (Static Method)
* 定義在類別內、但不需要存取類別（class）或實例（instance）屬性的獨立函式
* 靜態方法單純只是把一個功能邏輯打包在類別的命名空間下，方便管理

````python
class Book:
    # 類別變數
    count = 0

    # 類別方法，cls 可改其他名稱
    @classmethod
    def show_count(cls):
        print("cls.count:", cls.count)

    def __init__(self, isbn, name, price, author):
        Book.count += 1
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    # 實例方法，self 可改其他名稱
    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")

    # 靜態方法，@staticmethod 雖可不加，但建議加上
    @staticmethod
    def show_info(*books):
        for book in books:
            book.show()


def main():
    Book.show_count()
    print("-" * 30)
    book1 = Book("111", "Python", 500, "Paul")
    book2 = Book("222", "C++", 400, "Cindy")

    Book.show_info(book1, book2)

    print("-" * 30)
    Book.show_count()


main()
````

### 篩選
可使用 filter + lambda，但優先使用串列生成式 (list comprehension)，因為較 filter + lambda 執行效能更佳，可讀性也更高



````python
class Book:
    def __init__(self, isbn, name, price, author):
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")

    @staticmethod
    def show_books(books):
        for book in books:
            book.show()


# 測試資料
books = [
    Book("111", "Python", 500, "Paul"),
    Book("222", "C++", 400, "Cindy"),
    Book("333", "MongoDB", 600, "Mike")
]
print("測試資料:")
Book.show_books(books)
print("-" * 30)

print("篩選:")
keyword = input("搜尋書名: ")
# 優先使用串列生成式 (list comprehension)
# 因為較 filter + lambda 執行效能更佳，可讀性也更高
result = [book for book in books if keyword.lower() in book.name.lower()]

# filter + lambda
# result = list(
#     filter(lambda book: keyword.lower() in book.name.lower(), books)
# )

Book.show_books(result)
````

### 取最大值
max() + key

````python
class Book:
    def __init__(self, isbn, name, price, author):
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")

    @staticmethod
    def show_books(books):
        for book in books:
            book.show()


# 測試資料
books = [
    Book("111", "Python", 500, "Paul"),
    Book("222", "C++", 400, "Cindy"),
    Book("333", "MongoDB", 600, "Mike")
]
print("測試資料:")
Book.show_books(books)
print("-" * 30)

print("最大值 (最貴的書):")
max_book = max(books, key=lambda book: book.price)
max_book.show()
````

## 繼承
透過繼承 (inheritance) ，子類別 (derived / sub class) 可以重複利用父類別 (base / super class) 定義的內容 (屬性與方法)


````python
class Book:
    def __init__(self, isbn, name, price, author):
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")


# ComputerBook 繼承 Book
class ComputerBook(Book):
    # 子類別增加 example 屬性
    def __init__(self, isbn, name, price, author, example):
        # 呼叫父類別建構式
        super().__init__(isbn, name, price, author)
        self.example = example


def main():
    python = ComputerBook("111", "Python", 500, "Paul", "www.examples.com")
    python.show()
    print("example:", python.example)


main()
````

## 覆寫
如果父類別方法內容不適用於子類別，子類別可以改變內容以符合需求，稱作覆寫 (override)

````python
class Book:
    def __init__(self, isbn, name, price, author):
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")


# ComputerBook 繼承 Book
class ComputerBook(Book):
    # 子類別增加 example 屬性
    def __init__(self, isbn, name, price, author, example):
        # 呼叫父類別建構式
        super().__init__(isbn, name, price, author)
        self.example = example

    # override
    def show(self):
        super().show()
        print("example:", self.example)


def main():
    python = ComputerBook("111", "Python", 500, "Paul", "www.examples.com")
    python.show()


main()
````

## 物件的搜尋與排序
list.index() 可用於自訂物件的搜尋上，但需要覆寫 `__eq__()` 以指定欲比對的屬性

list.sort() 可用於自訂物件的排序上，但必須指定欲排序的屬性

````python
class Book:
    def __init__(self, isbn, name, price):
        self.isbn = isbn
        self.name = name
        self.price = price

    def __eq__(self, other):
        return self.isbn == other.isbn

    def __repr__(self):
        return f"{self.isbn}, {self.name}, {self.price}"

    @staticmethod
    def show_books(books):
        for book in books:
            print(book)

def main():
    # 測試資料
    books = [
        Book("111", "Python", 500),
        Book("222", "C++", 400),
        Book("333", "JS", 500)
    ]

    print("搜尋物件，列印所在索引: ")
    # 欲搜尋的對象
    js_book = Book("333", "JS", 500)
    # 搜尋自訂物件需要覆寫 __eq__() 以指定欲比對的屬性
    print("index:", books.index(js_book))
    print("-" * 30)

    print("價格降冪排序:")
    # 可以使用 lambda 指定要比對的屬性，預設為升冪，reverse=True 則為降冪
    books.sort(key=lambda book: book.price, reverse=True)
    Book.show_books(books)

main()
````

## 何謂值相同？
Set 特色為不允許元素值重複；也就是不允許值相同，否則無法加入，但何謂「值相同」？

物件欲存放至 set，所屬類別必須覆寫 object 類別的 `__eq__()` 與 `__hash__()` 方法
* `__eq__()`：明定何謂值相同

* `__hash__()`：可以實現搜尋效率極高的雜湊搜尋法


圖解雜湊搜尋法

<img src="https://drive.google.com/uc?export=view&id=1DKL-vVO4uN5346tUT3o3ft65eVGwayhl" width="500">


````python
class Book:
    def __init__(self, isbn, name, price):
        self.isbn = isbn
        self.name = name
        self.price = price

    def __eq__(self, other):
        return self.isbn == other.isbn

    def __hash__(self):
        return hash(self.isbn)

    def __repr__(self):
        return self.name


def main():
    books = set()
    books.add(Book("111", "MySQL", 600))
    books.add(Book("222", "Python", 500))
    books.add(Book("222", "Python", 500))
    print(books)


main()
````

## Set 數學集合功能

Set 內容為數列可以使用數學集合函式功能

Set 內容為物件一樣可以使用數學集合函式功能，但該物件所屬類別需要覆寫 `__eq__()` 與 `__hash__()` 方法

````python
class Book:
    def __init__(self, isbn, name, price):
        self.isbn = isbn
        self.name = name
        self.price = price

    def __eq__(self, other):
        return self.isbn == other.isbn

    def __hash__(self):
        return hash(self.isbn)

    def __repr__(self):
        return self.name


def main():
    bookshop1 = set()
    bookshop1.add(Book("111", "Python", 500))
    bookshop1.add(Book("222", "MySQL", 600))
    bookshop1.add(Book("333", "MongoDB", 680))

    bookshop2 = set()
    bookshop2.add(Book("111", "Python", 500))
    bookshop2.add(Book("444", "OpenAI API", 650))
    bookshop2.add(Book("555", "Deep Learning", 880))

    print("bookshop1:", bookshop1)
    print("bookshop2:", bookshop2)
    print("intersection:", bookshop1.intersection(bookshop2))
    print("union:", bookshop1.union(bookshop2))


main()
````

## 類型檢查
建立一個可能會儲存不同類型物件的 list，例如購物車

從 list 取物件出來必須使用「isinstance」檢查型別，方能呼叫該類型定義的方法。但類型眾多時會很繁瑣

````python
class Book:
    def __init__(self, id, name, price, author):
        self.id = id
        self.name = name
        self.price = price
        self.author = author

    def show_book(self):
        print("ISBN:", self.id)
        print("書名:", self.name)
        print("定價:", self.price)
        print("作者:", self.author)


class Toy:
    def __init__(self, id, name, price, maker):
        self.id = id
        self.name = name
        self.price = price
        self.maker = maker

    def show_toy(self):
        print("玩具編號:", self.id)
        print("玩具名:", self.name)
        print("定價:", self.price)
        print("製造商:", self.maker)


def main():
    cart = [
        Book("111", "Python", 500, "Paul"),
        Toy("t01", "Hello Kitty", 1200, "Kitty Inc.")
    ]

    print("購物車資訊如下：")
    for item in cart:
        print("-" * 30)
        # 檢查 item 是否屬於 Book
        if isinstance(item, Book):
            item.show_book()
        elif isinstance(item, Toy):
            item.show_toy()
        else:
            pass


main()
````

## 多型 (Polymorphism)
* 定義一個所有物件的共同父類型
* 建立各個子類型並覆寫父類型的方法
* 執行時會依據物件真實類型呼叫對應的覆寫方法，稱為多型呼叫 (polymorphic call)

````python
class Product:
    def __init__(self, id, name, price):
        self.id = id
        self.name = name
        self.price = price

    def show(self):
        print(self.id, self.name, self.price, sep=", ")


class Book(Product):
    def __init__(self, id, name, price, author):
        super().__init__(id, name, price)
        self.author = author

    def show(self):
        self.show_book()

    def show_book(self):
        print("ISBN:", self.id)
        print("書名:", self.name)
        print("定價:", self.price)
        print("作者:", self.author)


class Toy(Product):
    def __init__(self, id, name, price, maker):
        super().__init__(id, name, price)
        self.maker = maker

    def show(self):
        self.show_toy()

    def show_toy(self):
        print("玩具編號:", self.id)
        print("玩具名:", self.name)
        print("定價:", self.price)
        print("製造商:", self.maker)


def main():
    cart = [
        Book("111","Python", 500, "Paul"),
        Toy("t01", "Hello Kitty", 1200, "Kitty Inc.")
    ]
    print("購物車資訊如下：")
    for product in cart:
        print("-" * 30)
        product.show()


main()
````

## 抽象類別 (Abstract Class)
* 建立方式：繼承 ABC 類別以建立抽象類別
* 包含抽象方法 (abstract method)：抽象方法只定義架構，沒有內容，建議加上「@abstractmethod」標示為抽象方法
* 抽象類別基本上無法實例化
* 子類別需要實作 (implement) 抽象父類別的所有抽象方法
否則該子類別也無法實例化

````python
from abc import abstractmethod, ABC


# 透過繼承 ABC 類別建立抽象類別
class Sale(ABC):
    def __init__(self, discount):
        self.discount = discount

    # 抽象方法
    @abstractmethod
    def salePrice(self): pass

    # 抽象方法
    @abstractmethod
    def show(self): pass


class Book(Sale):
    def __init__(self, discount, isbn, name, price, author):
        super().__init__(discount)
        self.isbn = isbn
        self.name = name
        self.price = price
        self.author = author

    def salePrice(self):
        return self.price * (1 - self.discount)

    def show(self):
        print(self.isbn, self.name, self.price, self.author, sep=", ")


class Toy(Sale):
    def __init__(self, discount, id, name, price, maker):
        super().__init__(discount)
        self.id = id
        self.name = name
        self.price = price
        self.maker = maker

    def salePrice(self):
        return self.price * (1 - self.discount)

    def show(self):
        print(self.id, self.name, self.price, self.maker, sep=", ")


def main():
    cart = [
        Book(0.1, "111", "Python", 500, "Paul"),
        Toy(0.3, "t01", "Hello Kitty", 1200, "Kitty Inc.")
    ]

    print("購物車資訊如下：")
    total = 0.0
    for item in cart:
        total += item.salePrice()
        item.show()

    print("總金額 (已折扣)", total)


main()
````

## 抽象類別應用
抽象類別 (abstract class) 可以說是藍圖、規範，用以約束各個子類別要實作必要功能。

如果不訂規範就沒有約束力，各個開發者各自為政，造成模組使用者困擾。例如：
* A 負責 MySQL 模組，建立新增資料的函式為 add()
* B 負責 MongoDB 模組，建立儲存資料的函式為 insert()
* 模組使用者 (呼叫者) 就會非常麻煩，因為即使操作相同，但對不同資料庫操作的函式名稱卻不相同，如果更換資料庫，會造成原本開發好的程式需要大部分重寫

DAO (Data Access Object，資料存取物件) 以抽象類別建立，可約束各個 DAO 模組開發者遵守規範，避免各自為政，降低系統整合與維護的困難，並且實現高彈性的模組化設計

````python
from abc import abstractmethod, ABC


class Dao(ABC):
    @abstractmethod
    def add(self): pass

    @abstractmethod
    def update(self): pass

    @abstractmethod
    def delete(self): pass

    @abstractmethod
    def find(self): pass


class DaoMySql(Dao):
    def add(self):
        print("MySQL 新增功能")

    def update(self):
        print("MySQL 修改功能")

    def delete(self):
        print("MySQL 刪除功能")

    def find(self):
        print("MySQL 查尋功能")


class DaoMongoDB(Dao):
    def add(self):
        print("MongoDB 新增功能")

    def update(self):
        print("MongoDB 修改功能")

    def delete(self):
        print("MongoDB 刪除功能")

    def find(self):
        print("MongoDB 查尋功能")


def main():
    dao: Dao

    # 更改資料庫只要切換不同功能類別
    dao = DaoMySql()
    # dao = DaoMongoDB()

    # 以下程式無需更改
    dao.add()
    dao.update()
    dao.delete()
    dao.find()


main()
````

## 封裝
封裝（encapsulation）是把「資料 + 操作資料的方法」包在一起，並控制外部如何使用它。

封裝的重點
* 保護資料不被亂改
* 控制資料的存取方式
* 將複雜操作包起來，只讓使用者看到「該用的功能」

Python 的三種存取層級
* Public (公開)：外部存取：完全自由。屬性與方法預設為 public
* Protected (保護)：名稱前要加上 "_" (1 個底線)。供內部使用，外部不應直接修改
* Private (私有)：名稱前要加上 "__" (2 個底線)。Python 會自動將其更名成「_ClassName_PropertyName」 (例如：_Book_price) ，以降低誤用機率

Python 採取「信任開發者」的哲學：如果不希望外部去改，會採用 protected，但卻不硬性阻止。特殊情況才會使用 private。

封裝步驟
* 將屬性標示為 protected，代表外部不應直接修改
* 使用「@property_name.setter」建立 setter 處理存值操作
* 使用「@property」建立 getter 處理取值操作

````python
class Book:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # getter
    @property
    def name(self):
        return self._name

    # setter（加入檢查）
    @name.setter
    def name(self, name):
        if name.isspace():
            raise ValueError("name 不可為空")

        self._name = name

    # getter
    @property
    def price(self):
        return self._price

    # setter
    @price.setter
    def price(self, price):
        if price < 0:
            raise ValueError(f"price 不可為負數: {price}")

        self._price = price



def main():
    python = Book("Python", 500)
    # 呼叫 getter
    print("name:", python.name)
    print("price:", python.price)

    # price 錯誤
    # python = Book("Python", -500)

    # name 錯誤
    # python = Book(" ", 500)


main()
````
