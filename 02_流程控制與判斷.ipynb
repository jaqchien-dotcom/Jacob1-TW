## if-else 敘述句
* if-else 是用來做「條件判斷」的語法，讓程式根據不同情況執行不同程式碼
* 可以加入 elif (else if) 來增加需要判斷的條件
* if-else 區塊內可以再置入 if-else 區塊，而形成所謂的 if-else 巢狀架構




```python
# if-elif-else 敘述句
score = 87
text = f"{score}分，"

if score >= 85:
    text += "贏得10,000元獎學金"
elif score >= 75:
    text += "贏得5,000元獎學金"
else:
    text += "再努力"

print(text)
```

```python
# if-else 巢狀敘述句
score = 87
text = f"{score}分，"

if score >= 75:
    if score >= 85:
        text += "贏得10,000元獎學金"
    else:
        text += "贏得5,000元獎學金"
else:
    text += "再努力"

print(text)
```

## 條件運算式
* 邏輯觀念與 if-else 敘述句相同
* 條件運算式運算後會直接得到值，所以是運算式 (expression)

```python
score = 87
text = f"{score}分，"
text1 = "贏得10,000元獎學金！\n"
text2 = "再努力！\n"

text += text1 if score >= 85 else text2

print(text)
```

## match-case 條列式比對
* Python 3.10 開始支援 match-case
* 適合用於列舉資料上
* 由上至下順序與 case 的值比對是否相同

```python
rating = 5
match rating:
    case 1:
        print("第1名，10,000元獎金")
    case 2:
        print("第2名，5,000元獎金")
    case 3:
        print("第3名，2,000元獎金")
    case 4 | 5 as number:
        print(f"佳作(第{number}名)，500元獎金")
    case _:
        print("還要再努力")
```

## while 迴圈
while 迴圈條件式為 True 時會執行該迴圈區塊的程式內容，直到條件式為 False，才會結束 while 迴圈

但要小心別陷入無限迴圈！

```python
i = 1
while i <= 10:
    print(i, end=" ")
    i += 1

print("大於 10 即停止")
```

## for 迴圈
for 可搭配 range() 函式設定控制變數的初始值、終止值和變化量

```python
# 初始值為 1，終止值為 11 (到 11 就停止，所以不會印出 11），變化量為 +1
for i in range(1, 11, 1):
    print(i, end=" ")
```

```python
# 初始值為 1，終止值為 11，變化量預設為 +1
for i in range(1, 11):
    print(i, end=" ")
```

```python
# 初始值預設為 0，終止值為 11，變化量預設為 +1
for i in range(11):
    print(i, end=" ")
```

```python
# 初始值為 10，終止值為 0（到 0 就停止，所以不會印出 0），變化量為 -1
for i in range(10, 0, -1):
    print(i, end=" ")
```

## 巢狀迴圈
迴圈內再加入迴圈

```python
for i in range(5):
    print(f"i = {i}")
    for j in range(1, 11):
        print(f" {j}", end=" ")
    print()
```

## 特殊流程處理
if + break 可以在特定條件時強制結束迴圈

if + continue 可以強制結束該次迴圈執行，直接進入下次迴圈執行

```python
# if + break
floorStop = 5
for floor in range(1, 11):
    if floor >= floorStop:
        print(f"停在{floor}樓")
        break
    print(f"{floor}樓", end=" ")
```

```python
# if + continue
floorSkip = 4
for floor in range(1, 11):
    if floor == floorSkip:
        continue
    print(f"{floor}樓", end=" ")
```
