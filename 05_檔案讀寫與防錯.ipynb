## 存取文字檔案 - 簡易型
Python 存取文字檔最簡易的方式為：pathlib.Path 搭配 write_text() 與 read_text()

```python
from pathlib import Path

# 取得目前工作目錄
current_dir = Path().cwd()
print("目前工作目錄:", current_dir)
print()

# 建立存檔目錄 - test
test = current_dir/"test"
# parents=True 路徑中所有不存在的目錄都會建立
# exist_ok=True 目錄不存在就建立
test.mkdir(parents=True, exist_ok=True)

# 檔案路徑
file_path = test/"simple.txt"

# 檔案內容
text = "春暖花開"

# 存檔 - 簡易型
print("寫入文字 ...")
file_path.write_text(text, encoding="utf-8")
print("存檔完畢!\n")

# 讀檔 - 簡易型
print("讀取內容如下：")
print(file_path.read_text(encoding="utf-8"))
```

## 存取文字檔案 - 一般型
Python 存取文字檔也可使用 pathlib.Path 搭配 open() 方式，相關設定如下：
* "r"：讀取模式 (預設)。打開檔案用於讀取內容，如果檔案不存在則會引發FileNotFoundError
* "w"：寫入模式。如果檔案存在則會被覆蓋，如果檔案不存在則會建立新檔案
* "a"：附加模式。打開檔案並附加內容，如果檔案不存在則會建立新檔案
* "t"：文字模式 (預設)，用於讀寫文字檔案
* "b"：二進位模式，通常用於讀寫非文字檔案

```python
from pathlib import Path

# 建立存檔目錄 - test
test = Path().cwd()/"test"
test.mkdir(parents=True, exist_ok=True)

# 檔案路徑
file_path = test/"books.txt"

# 檔案內容
text = """
isbn, name, price
111, Python, 500
222, SQL, 600
333, Android App, 880
444, iOS App, 900
""".strip()

# 存檔
print("寫入文字 ...")
# "w" (write) 代表寫入資料，如果為文字可省略 "t" (text)
with file_path.open(mode="wt", encoding="utf-8") as f:
    f.write(text)

print("存檔完畢!\n")

# 讀檔
print("讀取內容如下：")
with file_path.open(mode="r", encoding="utf-8") as f:
    for line in f:
        print(line, end="")
```

## 存取 CSV 檔案
Python 存取 CSV 檔案最常使用內建的 csv 模組，不需要額外安裝第三方套件

```python
from pathlib import Path
import csv

# 建立存檔目錄 - test
test = Path().cwd()/"test"
test.mkdir(parents=True, exist_ok=True)

# 檔案路徑
file_path = test/"books.csv"

# 檔案內容
data = [
    {"isbn": "111", "name": "Python", "price": 500},
    {"isbn": "222", "name": "SQL", "price": 600},
    {"isbn": "333", "name": "Android App", "price": 880},
    {"isbn": "444", "name": "iOS App", "price": 900},
]

# 存成 CSV 檔案
print("存成 CSV 檔案 ...")
# fieldnames = data[0].keys():
# 從第一個 dictionary 中取出所有的 key，這些 key 將作為 CSV 檔案的標題列 (headers)
fieldnames = data[0].keys()

# newline="" 對於 CSV 寫入非常重要，可以防止某些作業系統產生額外的空行
with file_path.open("w", newline="", encoding="utf-8") as f:
    # 建立一個 csv.DictWriter 物件。這個物件專門將 dictionary 寫入 CSV 檔案
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)

print("CSV 存檔完畢!\n")

# 讀取 CSV 檔案
print("讀取 CSV 檔案內容如下：")
with file_path.open(encoding="utf-8") as f:
    # 將 CSV 每一列讀成一個 dictionary，欄位名稱來自第一列標題
    reader = csv.DictReader(f)
    data = list(reader)
    print(data)
```

## 存取 JSON 檔案
Python 處理 JSON 通常使用內建的 json 模組，不需要額外安裝第三方套件

```python
from pathlib import Path
import json

# 建立存檔目錄 - test
test = Path().cwd()/"test"
test.mkdir(parents=True, exist_ok=True)

# 檔案路徑
file_path = test/"books.json"

# 檔案內容
data = [
    {"isbn": "111", "name": "Python", "price": 500},
    {"isbn": "222", "name": "SQL", "price": 600},
    {"isbn": "333", "name": "Android App", "price": 880},
    {"isbn": "444", "name": "iOS App", "price": 900},
]

# 存成 JSON 檔案
print("存成 JSON 檔案 ...")
with file_path.open("w", encoding="utf-8") as f:
    # ensure_ascii=False 以確保非 ASCII 字 (例如中文字) 會以原始形式寫入
    # indent=2：設置縮排為 2 個空格，讓輸出的 JSON 檔案具有易於閱讀的格式
    json.dump(data, f, ensure_ascii=False, indent=2)

print("JSON 存檔完畢!\n")


# 讀取 JSON 檔案
print("讀取 JSON 檔案內容如下：")
with file_path.open("r", encoding="utf-8") as f:
    data = json.load(f)
    print(data)
```

## 執行錯誤
讀檔案可能發生執行錯誤 (exception)；如果不處理，Python 會：
* 立即停止目前執行流程
* 顯示錯誤類型與錯誤訊息 (Traceback)
* 強制結束程式

例如下列程式因為 data.txt 檔案不存在，會：
* 發生檔案不存在錯誤 (FileNotFoundError) 並立即停止目前執行流程
* 顯示錯誤行號與錯誤原因，但一般使用者無法輕易理解
* 強制結束程式，所以不會列印「讀取完成!」資訊

```python
from pathlib import Path

file_path = Path("data.txt")

print("讀取內容如下：")
with file_path.open("r", encoding="utf-8") as f:
    for line in f:
        print(line, end="")

print("讀取完成!")
```

## 錯誤處理
發生執行錯誤時，如果可以，應該予以適當處理，以達到下列目的：
* 顯示易懂的錯誤訊息，而非系統技術性的錯誤訊息，使用者方能明瞭錯誤原因
* 試圖將程式導正，例如：使用者輸入錯誤，讓他再重新輸入一次

錯誤處理機制
* try：try 區塊放可能會產生 exception 的程式碼
* except：可能有多個，將欲檢測的錯誤放在 except 後方，並將處理程式放在 except 區塊，加上 as 可以取得觸發錯誤的 exception 物件
* else：else 區塊放沒有發生 exception 時會執行的內容
* finally：finally 區塊放無論是否發生 exception 都要執行的內容 (例如釋放資源)

```python
def divide(dividend, divisor):
    return dividend / divisor

while True:
    try:
        dividend = float(input("請輸入被除數: "))
        divisor = float(input("請輸入除數: "))
        result = divide(dividend, divisor)
    except ValueError as err:
        # print(err) # 列印原始錯誤訊息
        print("數字格式錯誤，請再輸入一次")
        continue
    except ZeroDivisionError:
        print("除數不可為 0，請再輸入一次")
        continue
    else:
        print("運算結果:", result)
        break
    finally:
        print("finally 區塊")
```

## 手動釋放資源
程式使用資源時，無論程式處理過程成功或失敗，都應該關閉以釋放 (clean up) 該資源。

如果要手動釋放資源，釋放的程式碼需要放在 finally 區塊。


```python
from pathlib import Path

file_path = Path("test", "books.txt")
f = None

try:
    f = file_path.open("r", encoding="utf-8")
    print("讀取內容如下：")
    for line in f:
        print(line, end="")
    print()
except FileNotFoundError:
    print("檔案不存在")
except UnicodeDecodeError:
    print("檔案編碼錯誤")
finally:
    if f is not None:
        # 手動釋放檔案資源
        f.close()
        print("檔案已經關閉")
```

## 自動釋放資源
Python 已經支援自動釋放資源：
* 加上 with 敘述句即可在使用資源物件 (例如檔案) 時，無論是否發生 exception，都能自動釋放該資源
* 建議採用自動釋放資源寫法，可避免忘記關閉資源，而且可以減少撰寫 finally 區塊的麻煩

```python
from pathlib import Path

file_path = Path("test", "books.txt")

try:
    # 會自動釋放檔案資源
    with file_path.open("r", encoding="utf-8") as f:
        print("讀取內容如下：")
        for line in f:
            print(line, end="")
        print()
        print("\nf.closed (with內):", f.closed)

    print("f.closed (with外):", f.closed)
except FileNotFoundError:
    print("檔案不存在")
except UnicodeDecodeError:
    print("檔案編碼錯誤")
```
