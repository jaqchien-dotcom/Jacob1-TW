## 什麼是函式？
* 函式有名稱、參數 (parameter)、回傳值 (return value)
* 自成一個程式區塊 (block)
* 執行特定功能或特定計算

```python
# 無參數、無回傳值
def say_hello():
    print("Hello")

say_hello()
```

```python
# 2 個參數，無回傳值
def show_info(name, price):
    print(f"書名: {name}\n定價: {price}")

show_info("Python", 500)
```

```python
# 參數有預設值 (default arguments)
def show_book_info(name, price, discount=0.2):
    info = f"書名: {name}\n售價: {price * (1 - discount)}"
    print(info)

# 呼叫時不傳值代表使用預設值
show_book_info("Python", 500)

# 呼叫時也可傳值代表不使用預設值
show_book_info("Python", 500, 0.3)
```

```python
# 2 個參數與 1 個回傳值
def get_info(name, price):
    info = f"書名: {name}\n定價: {price}"
    return info

result = get_info("Python", 500)
print(result)
```

```python
# 多個參數與多個回傳值
def get_sale_info(name, price, discount):
    sale_price = price * (1 - discount)
    info = f"書名: {name}\n定價: {price}"
    return sale_price, info

# 使用索引取值
result = get_sale_info("Python", 500, 0.2)
print(f"{result[1]}\n特價: {result[0]}")
print("-" * 30)

# 使用變數取值 (tuple 會自動拆解)
sale_price, info = get_sale_info("MySQL", 800, 0.3)
print(f"{info}\n特價: {sale_price}")
```

```python
# Keyword argument
def show_book(name, price, discount):
    info = f"書名: {name}\n售價: {price * (1 - discount)}"
    print(info)

# 不指定名稱，需按照參數順序；指定名稱，可以不按照順序
show_book("Python", discount=0.2, price=500)
```

```python
# 「/」表示前面的參數只能使用 positional argument (位置參數) 傳入，不能使用關鍵字參數 (keyword argument)
# 「*」表示後面的參數必須使用 keyword argument (關鍵字參數) 傳入
def func(a, b, /, c, *, d, e):
    print(f"a={a}, b={b}, c={c}, d={d}, e={e}")

func(1, 2, 3, d=4, e=5)

# 錯誤，前三參數需為 positional argument
# func(a=1, b=2, c=3, d=4, e=5)

# 錯誤，後二參數需為 keyword argument
# func(1, 2, 3, 4, 5)
```

## 引數打包與拆解
呼叫函式支援 list、tuple、dictionary 引數自動打包 (packing) 與拆解 (unpacking)

打包成 tuple

```python
def pack_tuple(*numbers):
    print(f"{numbers}, sum = {sum(numbers)}")

pack_tuple(1, 2, 3, 4)
```

打包成 dictionary

```python
def pack_dict(**employee):
    print(employee)

pack_dict(name="John", age=30, city="Taipei")
```

將 list 拆解

```python
def unpack_list(n1, n2, n3):
    print(f"{n1} + {n2} + {n3} = {n1 + n2 + n3}")

list = [1, 2, 3]
unpack_list(*list)
```

將 dictionary 拆解

```python
def unpack_dict(name, age, city):
    print(f"name: {name}, age: {age}, city: {city}")

# dictionary 的 key 必須與參數名稱相同，否則錯誤
employee = {"name": "Ken", "age": 28, "city": "Taoyuan"}
unpack_dict(**employee)
```

## 函式可以當做參數


```python
# op_func 是函式型參數
def calculate(num1, num2, op_func):
    result = op_func(num1, num2)
    print("result:", result)


def divide(dividend, divisor):
    return dividend / divisor

calculate(10, 2, divide)
```

具體的應用範例

```python
# 連線的函式：連線成功即執行對應任務
def connect(url, task):
    print(f"連線到 {url} ...")
    # 假設連線成功
    success = True
    if success:
        print("連線成功")
        # 執行傳入的任務
        task()
    else:
        print("連線失敗")


# 定義函式：顯示商品資料
def show():
    print("讀取所有商品資料")
    print("顯示商品資料")

# 連線到指定網址，一旦連線成功，顯示商品資料
connect("www.shopping.com", show)
```

## Lambda
* lambda 是一個沒有名字、極其簡化的函式
* 必須是運算式，所以無需加上 return
* 常用於函式型參數的傳遞

```python
def calculate(num1, num2, op_func):
    result = op_func(num1, num2)
    print("result:", result)


# 定義函式並用於傳遞
def divide(dividend, divisor):
    return dividend / divisor

calculate(10, 2, divide)


# lambda 是個簡化版的函式
calculate(10, 2, lambda dividend, divisor: dividend / divisor)
```

## 標準函式庫 - 亂數函式
Python 的 random 模組用來產生隨機數、隨機選取資料以及打亂資料順序

```python
import random

print("產生 [1,100) 的整數亂數:")
print("random.randrange(1, 100):",
      random.randrange(1, 100))
print("-" * 30)

print("產生 [1,100) 的浮點亂數:")
print("random.uniform(1, 100):",
      random.uniform(1, 100))
print("-" * 30)

print("產生 6 個 [1,100) 可重複的整數亂數:")
print("random.choices(range(1, 100), k=6):\n",
      random.choices(range(1, 100), k=6))
print("-" * 30)

print("產生 6 個 [1,100) 不重複的整數亂數:")
print("random.sample(range(1, 100), k=6):\n",
      random.sample(range(1, 100), k=6))
print("-" * 30)

books = ["Python", "C++", "Java", "SQL"]
print("隨意挑出一個元素:")
print("random.choice(books):",
      random.choice(books))
print("-" * 30)

print("洗牌(隨機重排):")
random.shuffle(books)
print("random.shuffle(books):\n", books)
```

## Python 日期時間
Python 日期與時間功能主要由內建的 datetime 模組提供

datetime 類別專門處理日期與時間 (年、月、日、時、分、秒、微秒)，可精確到微秒 (microsecond, 10-6 秒)

```python
from datetime import datetime

print("現在日期時間:")
now = datetime.now()
print(now)

print("指定日期時間:")
dt = datetime(year=2000, month=1, day=1, hour=23)
print(dt)

print("-" * 30)

print("取得年、月、日、時、分、秒、微秒")
print("now.year:", now.year)
print("now.month:", now.month)
print("now.day:", now.day)
print("now.weekday():", now.weekday())  # Monday == 0 ... Sunday == 6.
print("now.hour:", now.hour)
print("now.minute:", now.minute)
print("now.second:", now.second)
print("now.microsecond:", now.microsecond)
```

date 類別專門處理日期 (年、月、日)，不包含時間資訊；它和 datetime 類別有許多相似的操作

```python
from datetime import date

print("今天日期:")
today = date.today()
print(today)
print("-" * 30)

print("指定日期:")
d = date(year=2000, month=1, day=1)
print(d)
```

## 日期時間格式化與解析

格式化：日期時間 → 字串，使用 strftime()

解析：字串 → 日期時間，使用 strptime()

自訂格式化與解析需要使用的 [日期時間格式符號](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes)

```python
from datetime import datetime

dt = datetime.now()
format = "%Y年%m月%d日 %H時%M分%S秒"

# 格式化 datetime 為 str
formatted = dt.strftime(format)
print("格式化結果:\n", formatted)

# 解析 str 為 datetime
parsed = dt.strptime(formatted, format)
print("解析結果:\n", parsed, type(parsed))
```

資料交換可使用 ISO 8601 格式

```python
from datetime import datetime

dt = datetime.now()

# 格式化 datetime 為 ISO 8601 格式
formatted = dt.isoformat()
print("ISO 8601 格式化結果:\n", formatted)

# 解析 ISO 8601 格式為 datetime
parsed = datetime.fromisoformat(formatted)
print("解析 ISO 8601 格式結果:\n", parsed, type(parsed))
```

## 時區
建立 datetime 物件沒有指定時區，就是建立當地日期時間

建立 datetime 物件可以利用 timedelta 指定時區

```python
from datetime import datetime, timedelta, timezone

# 無時區
now_local = datetime.now()
print("now_local:\n", now_local)

# 系統所在時區
print("now_local.astimezone():\n", now_local.astimezone())

# UTC時區
now_utc = datetime.now(timezone.utc)
print("now_utc:\n", now_utc)

# +8 時區
now_8h = datetime.now(timezone(timedelta(hours=8)))
print("now_8h:\n", now_8h)
```

建立具有日光節約時間資訊的 datetime 物件，需要使用 Python 3.9開始支援的 zoneinfo.ZoneInfo 來建立時區，這樣 Python 才會自動判斷是否正在實施日光節約時間

```python
from datetime import datetime
from zoneinfo import ZoneInfo

# 英國夏令開始：每年 3 月最後一個星期日；冬令開始：10 月最後一個星期日
# 使用 ZoneInfo 建立時區，才能支援日光節約時間
summer = datetime(
    2026, 7, 1,
    tzinfo=ZoneInfo("Europe/London")
)

winter = datetime(
    2026, 1, 1,
    tzinfo=ZoneInfo("Europe/London")
)

# 夏令與冬令時間的時差不同
print("夏令時間:", summer)
print("冬令時間:", winter)

now = datetime.now(ZoneInfo("Europe/London"))
# dst() 會回傳 timedelta；
# 夏令時間為 1:00:00 (表示有 1 小時 DST)；冬令時間則為 0:00:00
print("now.dst():", now.dst())
# 判斷是否為夏令時間
print("現在是夏令時間:", now.dst() != timedelta(0))
```

## 日期時間比較
可以直接使用比較運算符號：<, <=, >, >=, ==, !=


```python
from datetime import datetime

now = datetime.now()
birthday = datetime(year=2000, month=1, day=1, hour=23)
print("now > birthday:", now > birthday)
print("now == birthday:", now == birthday)
print("now < birthday:", now < birthday)
```

## 日期時間加減
主要透過 timedelta 類別完成

```python
from datetime import datetime, date, timedelta

now = datetime.now()
print("now:", now)

print("加減指定時間單位")
tdelta_1d_2h = timedelta(days=1, hours=2)
print("now + tdelta_1d_2h:", now + tdelta_1d_2h)
print("now - tdelta_1d_2h:", now - tdelta_1d_2h)

print("-" * 30)

print("計算兩日期差距")
d1 = date(2026, 6, 1)
d2 = date(2026, 6, 7)
delta = d2 - d1
print(f"{d2} - {d1} = {delta.days}")
```
