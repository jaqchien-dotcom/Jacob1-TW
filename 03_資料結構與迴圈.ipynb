## List
List 是一個放大量資料的容器，而且：
* 有順序（有 index）
* 可以重複
* 可以修改（mutable）
* 可以放不同型別

### 建立空的 list 並且加入元素

```python
names = []
# 下列方式也可建立空的 list
# names = list()

names.append("Python")
names.append("Java")

print(names)
```

### 建立 list 並同時給予初始值

```python
names = ["Python", "Java", "JS", "Swift", "C#"]
print("names:", names)

# 取得 list 長度
print("len(names):", len(names))
```

### 透過索引 (index) 取值

```python
names = ["Python", "Java", "JS", "Swift", "C#"]

# 取出索引為 0 的元素
print("names[0]:", names[0])

# 取後面數來第一個元素
print("names[-1]:", names[-1])

# 索引超過界線會產生錯誤 IndexError: list index out of range
# print("names[5]:", names[5])

# 取索引 0 (含) 到 3 (不含) 的元素
print("names[0:3]:", names[0:3])
# 開始索引不寫為 0
print("names[:3]:", names[:3])
# 結束索引不寫為直到最後元素
print("names[2:]:", names[2:])
```

### 搭配 for 迴圈遍歷資料

```python
names = ["Python", "Java", "JS", "Swift", "C#"]

for name in names:
    print(name, end=" ")
```

### List 內容異動

```python
names = ["Python", "Java", "JS", "Swift", "C#"]

# 設定值
names[0] = "Python 3"
print("修改索引 0 的值:", names)
print("-" * 30)

# 移除指定值
names.remove("Java")
print("移除 Java:", names)
print("-" * 30)

# pop()即為pop(-1)，移除末端值
names.pop()
print("移除末端值:", names)
print("-" * 30)

# 清空list內容
names.clear()
print("清空 list:", names)
```

### List 排序
將 list 內的元素依照一定順序排列

```python
names = ["Python", "Java", "JS", "Swift", "C#"]

# 升冪排序
names.sort()
print(f"升冪排序: {names}")

# 降冪排序
names.sort(reverse=True)
print(f"降冪排序: {names}")
```

### List 搜尋
在 list 內找到指定值並回傳對應結果

```python
numbers = [9, 1, 3, 7, 5, 6, 2, 8, 10, 4]

target = int(input("欲搜尋的值: "))

# 找到回傳 True
if target in numbers:
    print(f"{target} 存在")
else:
    print(f"{target} 不存在")

# 找到該元素回傳對應的索引，搜尋不到產生 ValueError (用 try-except 解)
firstIndex = numbers.index(target)
print(f"{target} 的索引為 {firstIndex}")
```

### 關鍵字搜尋
in 用於字串上，只要局部符合即可

```python
names = ["Python AI", "Java", "JS", "Swift", "C#"]

keyword = input("搜尋書名: ")
for name in names:
    # in 用於字串上，只要局部符合即可
    if keyword.lower() in name.lower():
        print(name)
```

## Dictionary
是一種以 key-value 方式儲存資料的資料結構

### 建立空的 dictionary 並且加入元素

```python
book = {}
# 下列方式也可建立空的 dictionary
# book = dict()

book["name"] = "Python"
book["price"] = 500
book["author"] = "Paul"

print(book)
```

### 建立 dictionary 並同時給予初始值

```python
book = {
    "name": "MySQL",
    "price": 800,
    "author": "Mike"
}

print(book)

# 取得 dictionary 長度
print("len(book):", len(book))
```

### 透過 key 取值

```python
book = {
    "name": "Python",
    "price": 500,
    "author": "Paul"
}

print(book["name"])
print(book["price"])
```

取得 keys 與 values

```python
book = {
    "name": "Python",
    "price": 500,
    "author": "Paul"
}

# 取得 keys
for key in book.keys():
    print(key)

print("-" * 30)

# 取得 values
for value in book.values():
    print(value)

print("-" * 30)

# 呼叫 items() 可同時取得 keys 與 values
for key, value in book.items():
    print(f"{key} - {value}")
```

### Dictionary 內容異動

```python
book = {
    "name": "Python",
    "price": 500,
    "author": "Paul"
}

# 改變指定 key 的 value
book["author"] = "Peter"
print("author 值改為 Peter:\n", book)
print("-" * 30)

# 新增 key - value
book["isbn"] = "123456789012"
print("新增 isbn 資料:\n", book)
print("-" * 30)

# 移除 key 代表的資料
book.pop("isbn")
print("移除 isbn 資料:\n", book)
print("-" * 30)

# 清空 dictionary 內容
book.clear()
print("清空 dictionary 內容:", book)
```

## Set
Set 容器內不允許元素值重複、而且不使用索引

```python
names = {"Python", "Java", "JS", "Swift", "C#"}

# 搭配 for 迴圈遍歷資料
for name in names:
    print(name, end=" ")
print()

# 欲加入的元素如果與 set 內既存的元素值相同，則無法加入
names.add("Python")
print(f"新增 Python: {names}")

names.remove("Swift")
print(f"移除元素 Swift: {names}")
```

### Set 與 List 可相互轉換

```python
names = {"Python", "Java", "JS", "Swift", "C#"}

# set 轉成 list
my_list = list(names)
print(f"set 轉成 list: {my_list}")

# list 轉成 set (list 元素值如有重複會被去除)
my_set = set(my_list)
print(f"list 轉成 set: {my_set}")
```

### Set 數學集合功能
Set 另外提供數學集合相關函式功能

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
c = {1, 2}
d = {1, 2}
print(f"a: {a}")
print(f"b: {b}")
print(f"c: {c}")
print(f"d: {d}")
print(f"a.intersection(b): {a.intersection(b)}")
print(f"a.union(b): {a.union(b)}")
print(f"a.symmetric_difference(b): {a.symmetric_difference(b)}")
print(f"c.issubset(a): {c.issubset(a)}")
print(f"c.issubset(d): {c.issubset(d)}")
print(f"a.issuperset(c): {a.issuperset(c)}")
print(f"b.isdisjoint(c): {b.isdisjoint(c)}")
print(f"c == d: {c == d}")
print(f"c is d: {c is d}")
```
