## Print 輸出方式
* 預設列印完畢會換行
* 列印完不想換行，使用 end 指定要取代換行的字串
* 列印多個參數可使用 sep (separator) 指定分隔符號

```python
num1 = 1
num2 = 2
sum1 = num1 + num2

# print() 列印完畢會換行
print(num1)
print(num2)
```

```python
# 列印完不想換行，使用 end 參數指定結束字串
print(num1, num2, end=", ")
print(sum1)
```

```python
# 列印多個參數可使用 sep (separator) 指定分隔符號，
# 沒有使用 end 參數則列印完後仍會換行
print(num1, num2, sep=", ")
print(sum1)
```

```python
# 既指定分隔符號，也指定列印完的結果
print(num1, num2, sep=" + ", end=" = ")
print(sum1)
```

## 文字類型
* str (String) 是 Python 的文字型別，用來儲存一個或多個字元，預設使用 Unicode
* 可以使用單引號 (single quotes) 或雙引號 (double quotes) 來包覆單行內容
「'Hello'」、「"Hello"」
* 可以使用三引號，「'''」(triple single quotes) 或「"""」(triple double quotes) 來包覆多行內容

```python
# 文字值可單或雙引號包覆
text1 = 'Hello, John.'
print(text1)

text2 = "Hello, John.\nHi, Mary."
print(text2)
```

```python
# 多行可三引號包覆，可讀性更高
text3 = """
Hello, John.
Hi, Mary.
"""
print(text3)
```

## 跳脫字元
是指在字串中具有特殊功能的符號，必須搭配反斜線 \ 使用，才能表示「特殊意義」或「字面字元」。常見跳脫字元如下：
* \n  換行 (line feed)
* \t  相當於按 tab 鍵 (horizontal tab)
* \r	 相當於按 return 鍵 (carriage return)
* \’	 取消單引號原有功能，成為單純的單引號文字 (single quote)
* \"	 取消雙引號原有功能，成為單純的雙引號文字 (double quote)
* \\	 取消反斜線原有功能，成為單純的反斜線文字 (backslash)

```python
# 換行
s1 = "書名：\nPython入門\n"
print(s1)
```

```python
# tab 鍵，上下欄位對齊
s2 = "Python入門\t650元\nMySQL資料庫管理\t700元\n"
print(s2)
```

```python
# 雙引號
s3 = "\"Python\"入門"
print(s3)
```

```python
# 反斜線
s4 = "\\Python\\入門"
print(s4)
```

## 文字串接
* 使用「+」進行串接 (Concatenation)：只能串接 字串與字串，不能 文字與數字
* f-strings (formatted string literals) 串接：Python 3.6 開始支援，是目前最常用、最方便的方法。在字串前加 f，並在 {} 中放入變數或運算式

```python
# 使用 + (加號) 串接
s1 = "Hello " + "World!"
print(s1)

a = 35
b = 20
# 將數字轉成文字後方能串接
s2 = "a = " + str(a) + ", b = " + str(b) + "; a x b = " + str(a * b)
print(s2)

# Python 3.6 開始支援 f-strings
s3 = f"a = {a}, b = {b}; a x b = {a * b}"
print(s3)
```

## 格式化
* 在程式設計中，格式化 (format) 是指控制輸出資料的呈現方式，讓資料更清楚、易讀、符合需求
* 可使用 f-strings 格式化

```python
num1 = 100000.0
num2 = 3.0

print("未格式化:")
print(str(num1) + " / " + str(num2) + " = " +  str(num1 / num2))

print("使用 f-strings 格式化:")
print(f"{num1:,.0f} / {num2:,.0f} = {num1 / num2:,.2f}")
```

## 數字類型與算術運算符號
整數類型：int

浮點數類型：float

常用算術運算符號
* +：  加 (addition)  
* -：  減 (subtraction)
* *：  乘 (multiplication)
* /：  除法 - 取商 (division)
* //：  整數除法 - 取商 (integer division，商為整數，不為小數)  
* %：  除法 - 取餘數 (remainder)
* **：  次方


```python
i1 = 5
i2 = 2
d1 = 2.0

# 加減乘除
print(f"{i1} + {i2} = {i1 + i2}")
print(f"{i1} - {i2} = {i1 - i2}")
print(f"{i1} * {i2} * {d1} = {i1 * i2 * d1}")
print(f"{i1} / {i2} = {i1 / i2}")

# 整數除法 (商為整數，不為小數)
print(f"{i1} // {i2} = {i1 // i2}")

# 取餘數
print(f"{i1} % {i2} = {i1 % i2}")

# 次方
print(f"{i1} ** {i2} = {i1 ** i2}")
```

## 複合指派運算符號
指派運算符號可以和其他運算符號結合，成為複合指派運算符號 (compound assignment operator)
* += ：   x += y 相當於 x = x + y
* -=  ：   x -= y 相當於 x = x - y
* *=  ：   x *= y 相當於 x = x * y
* /=  ：   x /= y 相當於 x = x / y
* %=：   x %= y 相當於 x = x % y

```python
a = 1
a += 1
print(f"a: {a}")
```

## 類型檢查與轉換
type() 用於檢查型別

整數與浮點數運算會自動轉為浮點數

型別轉換函式
* int(value) 用於轉成整數
* float(value) 用於轉成浮點數
* str(value) 用於轉成文字

```python
i = 1
f = 11.1

# type() 用於檢查型別
print("type(i):", type(i))
print("type(f):", type(f))

# 整數與浮點數運算會自動轉為浮點數
sum1 = i + f
print("sum1:", sum1)

# int() 用於轉成整數，float() 用於轉成浮點數
num1 = int("1")
num2 = float("11.1")
sum2 = int(num1 + num2)
print("sum2:", sum2)

# str() 用於轉成文字
text = "小華" + str(20) + "歲"
print(text)
```

## 布林類型
布林類型 (bool)，只有下列 2 種值
* True (成立)
* False (不成立)

比較運算符號 (comparison operator) 如下，比較結果為布林類型
* 「>」： 大於 (greater than, a > b)
* 「>=」： 大於或等於 (greater than or equal to, a >= b)
* 「<」： 小於 (less than, a < b)
* 「<=」： 小於或等於 (less than or equal to, a <= b)
* 「==」： 等於 (equal to, a == b)
* 「!=」： 不等於 (not equal to, a != b)

邏輯運算符號

AND 運算符號
* 所有條件為 True，最後結果才會為 True
* AND 速算 (short-circuit evaluation)：
只要任一條件為 False，其後條件式皆不予判斷，結果直接為 False

```python
# 需要 100,000 元以及 10 天的假期才能參加澳洲旅行團
money = int(input("你有多少錢："))
day = int(input("你有幾天假："))
travel = money >= 100000 and day >= 10
if travel:
    print("可以去澳洲看黃金海岸")
else:
    print("只能在公園看黃金獵犬")
```

OR 運算符號
* 只要有 1 個條件為 True，最後結果就為 True
* OR 速算：只要有任一條件為 True，直接得到結果為 True

```python
# 去遊玩搭乘的交通工具可以為飛機或船
flight_ticket = True
boat_ticket = True
result = flight_ticket or boat_ticket
print(f"是否成行: {result}")
```

NOT 運算符號
* 會讓布林值反轉，就是會讓 True 變成 False，False 變成 True
* 一般為了處理反向邏輯會使用 NOT

```python
# NOT 運算符號會讓布林值反轉
print(not True)
print(not False)

# 限制級電影，觀眾成年可以進入；未成年會受到阻攔
age = int(input("請輸入年齡："))
adult = age >= 18
if not adult:
    print("未成年不能看限制級電影！")
```
